```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import os,json
```

```python
df = pd.read_csv("../Data/Original Data/title_basics.csv.gz")
df.head()
```

```python
df["genres_split"] = df["genres"].str.split(",")
df.head()
```

```python
exploded_genres = df.explode("genres_split")
exploded_genres
```

```python
genres_split = df["genres"].str.split(",")
genres_split
```

```python
unique_genres = genres_split.explode().unique()
unique_genres
```

```python
unique_genres = sorted(exploded_genres["genres_split"].unique())
unique_genres
```

```python
title_genres = exploded_genres[["tconst", "genres_split"]].copy()
title_genres.head()
```

```python
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints)) 
genre_map
```

```python
title_genres["genre_id"] = title_genres["genres_split"].map(genre_map)
title_genres = title_genres.drop(columns = "genres_split")
```

```python
title_genres.head()
```

```python
genre_lookup = pd.DataFrame({"Genre_Name": genre_map.keys(), 
                            "Genre_ID": genre_map.values()})
genre_lookup.head()
```

### Creating Primary Keys - title_basics

```python
from sqlalchemy.types import *

key_len = basics_df['tconst'].fillna('').map(len).max()
title_len = basics_df['primaryTitle'].fillna('').map(len).max()

basics_df_schema = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'runtimeMinutes':Integer()}

```

```python
basics_df.to_sql('title_basics', engine, dtype= basics_df_schema, if_exists='replace',index=False)
```

```python
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
```

### Creating Primary Keys - title_ratings

```python
key_len_2 = ratings_df['tconst'].fillna('').map(len).max()

ratings_df_schema = {
    "tconst": String(key_len_2+1), 
    'averageRating':Float(),
    'numVotes':Integer()}
```

```python
ratings_df.to_sql('title_ratings', engine, dtype= ratings_df_schema, if_exists='replace',index=False)

engine.execute('ALTER TABLE title_ratings ADD PRIMARY KEY (`tconst`);')



```
